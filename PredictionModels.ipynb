```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error

%matplotlib inline
```

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
```

```python
from sklearn.linear_model import LogisticRegression
xtrain = df_train[["OverallQual"]] #, "BsmtFinSF1", "TotalBsmtSF","1stFlrSF", "2ndFlrSF", "GrLivArea"]]
ytrain = df_train["SalePrice"]

log_reg = LogisticRegression()
log_reg_fit = log_reg.fit(xtrain, ytrain)
```

```python
yhat_log_reg = log_reg.predict(xtrain)
mean_squared_error(ytrain, yhat_log_reg)
```

```python
plt.scatter(xtrain, ytrain, color="black")
plt.plot(xtrain, log_reg.predict(xtrain), color="blue")
```

```python
from sklearn.linear_model import LinearRegression
xtrain = df_train[["OverallQual"]] #, "BsmtFinSF1", "TotalBsmtSF","1stFlrSF", "2ndFlrSF", "GrLivArea"]]
ytrain = df_train["SalePrice"]

lin_reg = LinearRegression()
lin_reg_fit = lin_reg.fit(xtrain, ytrain)
```

```python
yhat_lin_reg = lin_reg.predict(xtrain)
mean_squared_error(ytrain, yhat_lin_reg)
```

```python
plt.scatter(xtrain, ytrain, color="black")
plt.plot(xtrain, lin_reg.predict(xtrain), color="blue")
```
