```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv("train.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().any()
```

```python
house_ids = df.index
sale_prices = df["SalePrice"]
plt.scatter(house_ids, sale_prices)
plt.xlabel("House ids")
plt.ylabel("Sale prices")
plt.title("Sale prices vs house ids")
```

```python
lot_areas = df["LotArea"]
plt.scatter(lot_areas, sale_prices)
plt.xlabel("Lot areas")
plt.ylabel("Sale prices")
plt.title("Sale prices vs lot areas")
```

```python
mssubclass = df["MSSubClass"]
plt.scatter(mssubclass, sale_prices)
plt.xlabel("MSSubClass")
plt.ylabel("House prices")
plt.title("House prices vs MSSubClass")
```

```python
# Get columns with no empty values
isnull_any_list = df.isnull().any()
df_no_null_values = df[isnull_any_list[1:].index[isnull_any_list[1:] == False]]
```

```python
df_no_null_values.shape
```

```python
# Get only columns of type int64
df_no_null_int64 = df_no_null_values[df_no_null_values.dtypes.index[df_no_null_values.dtypes == "int64"]]
```

```python
df_no_null_int64.shape
```

```python
fig, axes = plt.subplots(len(df_no_null_int64.columns.values)-1, 1, sharey=True)

for i, col in enumerate(df_no_null_int64.columns.values[:-1]):
    df_no_null_int64.plot(x=[col], y=["SalePrice"], kind="scatter", ax=axes[i], figsize=(10, 200))
```
